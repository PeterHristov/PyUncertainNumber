**get CDF curve**

- two ways to do this

```python
# eg. loc=1, scale=2

support_l = norm.ppf(q=0.01, loc=1, scale=2) 
support_r = norm.ppf(q=0.99, loc=1, scale=2) 
```

```python
x_values = np.linspace(support_l, support_r, 1000)
```

```python
p_values = norm.cdf(x_values, loc=1, scale=2)
```

```python
fig, ax = plt.subplots()
ax.plot(x_values, p_values, '-')
ax.set_title('First way of plotting CDF')
```

```python
''' now let's look at the second way '''

p_values = np.linspace(0, 1, 1000)
x_values = norm.ppf(p_values, loc=1, scale=2)
```
