# mixture of DS structures

```python
lower_endpoints = np.random.uniform(-0.5, 0.5, 7)
upper_endpoints = np.random.uniform(0.5, 1.5, 7)
m_weights = [0.1, 0.1, 0.25, 0.15, 0.1, 0.1, 0.2] 
```

```python
b = [pba.I(couple) for couple in zip(lower_endpoints, upper_endpoints)]
```

```python
b[2]
```

```python
ax = plot_intervals(b)
ax.margins(x=0.2, y=0.2)
ax.set_title('expert opinion in intervals')
```

```python
with mpl.rc_context({"text.usetex": True}):
    ax = plot_DS_structure(b, weights=m_weights)
    ax.margins(x=0.2, y=0.2)
    ax.set_yticks([])
    ax.set_title('Dempster Shafer structures')
```

```python
pbox_mix = mixture(*b, weights=m_weights)
axe = pbox_mix.display()
# axe.set_xlim([-0.5, 1.5])
axe.margins(x=0.2)
```
