```python
''' original ones from Scott '''

many = 2000

def sh(x, t, Data=None):
    """ plot CDF of both raw data and estimated dist constructs 

    args:
        x (array like): random deviates which approximates a distribution
        t (str): title
    """
    if Data is None:
        Data = data  # data mean the global var `data`
    plt.title(t)
    edf(Data, 'y')
    edf(x)
    plt.show()


def edf(d, c=None, lw=None, ls=None):
    """ plot CDF from either raw data or a repre of dist construct """
    
    if d.size == (2*many):  # p-box
        z, p = ecdf(d[0:many])
        plt.plot(z, p, c=c, lw=lw, ls=ls)
        z, p = ecdf(d[(many):(2*many)])
        plt.plot(z, p, c=c, lw=lw, ls=ls)
    else:  # distribution
        z, p = ecdf(d)
        plt.plot(z, p, c=c, lw=lw, ls=ls)
    # plt.ylabel('Cumulative probability') # just makes the graph smaller


def ecdf(d):
    """ prepare for empirical cdf from data 
    
    args:
        d (array): either sample data or deviates as approximated dist construct
    """
    
    d = np.array(d)
    N = d.size
    pp = np.concatenate((np.arange(N), np.arange(1, N+1)))/N
    dd = np.concatenate((d, d))
    dd.sort()
    pp.sort()
    return dd, pp
```

```python
''' scott's ramdom deviate-based implementation '''

def normal(m,s) : return(scipy.stats.norm.rvs(m,s, size=many))

# it returns random deviates generated from sample estimates
def MMgaussian(x) : return(normal(x.mean(), x.std())) 
```
